```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import pandas as pd
```

```python
training_ds = pd.read_csv('training.csv')
test_ds = pd.read_csv('test.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
sex = pd.get_dummies(training_ds['Sex'], drop_first = True)
embarked = pd.get_dummies(training_ds['Embarked'], drop_first = True)
Pclass = pd.get_dummies(training_ds['Pclass'], drop_first = True)
```

```python

training_ds = pd.concat([training_ds, sex, embarked,Pclass], axis = 1)
training_ds.head()
```

```python
training_ds.drop(['Sex', 'Pclass', 'Embarked', 'Cabin', 'Name', 'Ticket'],axis = 1, inplace = True)
training_ds.head()
```

```python
#check for null values
training_ds.isnull().sum()
```

```python
#handling null values
import numpy as np
train_values = {'Age' : round(np.mean(training_ds['Age']))}
training_ds = training_ds.fillna(value = train_values)
training_ds.head()



```

```python
sex = pd.get_dummies(test_ds['Sex'], drop_first = True)
embarked = pd.get_dummies(test_ds['Embarked'], drop_first = True)
Pclass = pd.get_dummies(test_ds['Pclass'], drop_first = True)
```

```python
test_ds = pd.concat([test_ds, sex, embarked,Pclass], axis = 1)
test_ds.head()
```

```python
test_ds.drop(['Sex', 'Pclass', 'Embarked', 'Cabin', 'Name', 'Ticket'],axis = 1, inplace = True)
test_ds.head()
```

```python
test_ds.isnull().sum()
```

```python
test_values = {'Age' : round(np.mean(test_ds['Age']))}
test_ds = test_ds.fillna(value = test_values)
test_ds.head()
```

```python
X = training_ds.drop('Survived', axis = 1)
y = training_ds['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=1)
```

```python
logmodel = LogisticRegression(solver = 'liblinear')
```

```python
logmodel.fit(X_train.values, y_train)
```

```python
LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=100,
                   multi_class='auto', n_jobs=None, penalty='l2',
                   random_state=None, solver='liblinear', tol=0.0001, verbose=0,
                   warm_start=False)
```

```python
predictions = logmodel.predict(X_test.values)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
print(classification_report(y_test, predictions))
```

```python
print(accuracy_score(y_test, predictions))
```

```python
final_test_predictions = logmodel.predict(test_ds.values)
```

```python
submissionfile = pd.read_csv('newsubmission.csv')
submissionfile['Survived'] = final_test_predictions
submissionfile.to_csv('submission.csv', index = False)
```
